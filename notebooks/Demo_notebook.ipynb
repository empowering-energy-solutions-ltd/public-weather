## Manually getting weather data for a test site

```python
path_analysis_results = Path.cwd().parent / 'data/'
simulation_year = 2020
demo_location = location.GeoLocation("Demo_site", 52.414, -1.143, 90.6,
                                     'UTC')
```

```python
era5_api = extractor.Era5DataExtractor(
        demo_location, saving_path=path_analysis_results)
ghi_data = era5_api.get_ghi_data(simulation_year)
temp_data = era5_api.get_temperature_data(simulation_year)
```

```python
demo_weather = weather_api.WeatherData(demo_location, simulation_year,weather_data_source=enums.WeatherDataSource.ERA5, saving_path=path_analysis_results)
```

#### Example data with solar components

```python
dataf = demo_weather.add_solar_components_to_ghi_data(ghi_data)
dataf
```

```python
weather_data = demo_weather.get_weather_data()
weather_data
```
